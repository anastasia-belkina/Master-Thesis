```python
import matplotlib.pyplot as plt
label_counts = df['Label'].value_counts()

print(label_counts)

label_counts.plot(kind='bar')
plt.xlabel('Label')
plt.ylabel('Count')
plt.title('Histogram of Label Counts')
plt.show()
```
