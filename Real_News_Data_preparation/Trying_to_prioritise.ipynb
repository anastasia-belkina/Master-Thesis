**Ставим эксперементы над самостоятельной разметкой NER**

```python
import spacy
import en_core_web_trf
nlp = spacy.load("en_core_web_trf")
```

```python
additional_terms = {"person", "man", "woman", "i", "me", "you", "he", "she", "him", "her", "they", "them", "himself", "herself", 
                    "police", "president", "minister", "administration", "immigrants"}
```

```python
# Define a function to find interacting entities based on the priority rules
def find_interacting_entities(doc):
    priority_labels = ["PERSON", "NORP", "ORG", "GPE"]
    valid_entities = []

    # Check entities in order of priority
    for label in priority_labels:
        if len(valid_entities) >= 2:
            break
        current_entities = [(ent.start_char, ent.end_char, ent.text) for ent in doc.ents if ent.label_ == label]
        valid_entities.extend(current_entities)
        if label == "PERSON":
            if len(valid_entities) >= 2: break
            else: 
                # Add tokens that are in the additional terms list
                for token in doc:
                    if token.text.lower() in additional_terms:
                        valid_entities.append((token.idx, token.idx + len(token.text), token.text))

        # Sort entities by their start position
        valid_entities = sorted(valid_entities, key=lambda x: x[0])
    
        # Merge adjacent entities
        merged_entities = []
        i = 0
        while i < len(valid_entities):
            current_ent = valid_entities[i]
            j = i + 1
            while j < len(valid_entities) and valid_entities[j][0] <= current_ent[1] + 1:  # Allow a gap of 1 character (space) between entities
                current_ent = (current_ent[0], valid_entities[j][1], current_ent[2] + " " + valid_entities[j][2])
                j += 1
            merged_entities.append(current_ent)
            i = j
        valid_entities = merged_entities
    
    # If we have at least two valid entities, return the first two
    if len(valid_entities) >= 2:
        return valid_entities[:2]
    else:
        return valid_entities
```

```python
def wrap_selected_entities(text):
    doc = nlp(text)
    entities = find_interacting_entities(doc)

    # Print the document if there are not exactly 2 entities
    if len(entities) != 2:
        print(f"Document with unexpected number of entities: {doc.text}")
    
    formatted_text = text
    offset = 0
    for ent in entities:
        start, end, ent_text = ent
        start += offset
        end += offset
        formatted_text = formatted_text[:start] + '[' + ent_text + ']' + formatted_text[end:]
        offset += 2  # 2 characters for the added brackets
    return formatted_text
```

```python
# Apply the function to the 'Sentence' column
df_test_shuffled_sen_lab['Formatted_Sentence'] = df_test_shuffled_sen_lab['Sentence'].apply(wrap_selected_entities)

# Create a new dataframe with the required columns
df_test_shuffled_sen_lab_sen_lab = df_test_shuffled_sen_lab[['Formatted_Sentence', 'Label']]

# Display the new dataframe
print(df_test_shuffled_sen_lab_sen_lab)
```
